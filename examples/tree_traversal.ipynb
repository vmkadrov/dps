```sage
var('xx,x,t')
a_param = var(['a'] +['a' + str(i) for i in range(10)])
f = function('f', nargs = 1)
F = a0*f(xx) + a1*f(x)
```

```sage
L = function('L') # оператор

F = a0*f(x)+ a1*L(f(x))*dt +\
    a2*f(x + a3*f(x)*dt + a4*L(f(x))*dt^2) +\
    a4*f(x + a5*f(x)*dt + a6*L(f(x))*dt^2 + a7*L(L(f(x)))*dt^3)
```
